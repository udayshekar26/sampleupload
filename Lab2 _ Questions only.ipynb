Using a function find the warmest and the coldest day of the week and the corresponding temperature. Ouput : Warmest is ['tue':34.5] Coldest is ['sat':25.5]


```
def maxmin(temps1):
  out = sorted(temps1, key = operator.itemgetter(1))
  return(out)


import operator
temps = [['sun',33.3], ['mon',32.1], ['tue',34.5], ['wed',28.7], ['thu',29.9], ['fri',28.6], ['sat',25.5]]
temps = maxmin(temps)
print("The Coldest Day is", temps[0])
print("The Warmest Day is", temps[len(temps) - 1])
```
